```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_neurons = 5
pre_stimulus_duration = 30  # seconds
post_stimulus_duration = 30  # seconds
baseline_rate = 10  # spikes per 30 seconds
excitation_factor = np.random.uniform(2, 3)  # 2-3x increase in spikes
inhibition_factor = np.random.uniform(0, 0.5)  # 0-0.5x decrease in spikes

def generate_spike_times(rate, duration):
    """Generates spike times for a given firing rate and duration."""
    num_spikes = np.random.poisson(rate)
    return np.sort(np.random.uniform(0, duration, num_spikes))

# Generate spike times for each neuron
neuron_spike_times = []
for neuron in range(num_neurons):
    pre_stimulus_spikes = generate_spike_times(baseline_rate, pre_stimulus_duration)
    if neuron % 2 == 0:  # Even neurons - inhibited
        post_stimulus_spikes = generate_spike_times(baseline_rate * inhibition_factor, post_stimulus_duration)
    else:  # Odd neurons - excited
        post_stimulus_spikes = generate_spike_times(baseline_rate * excitation_factor, post_stimulus_duration)
    
    spikes = np.concatenate((pre_stimulus_spikes, post_stimulus_spikes + pre_stimulus_duration))
    neuron_spike_times.append(spikes)

# Plot raster plot
plt.figure(figsize=(10, 6))
for neuron, spikes in enumerate(neuron_spike_times):
    plt.vlines(spikes, neuron + 0.5, neuron + 1.5)

# Add stimulus onset marker
plt.axvline(x=pre_stimulus_duration, color='red', linestyle='--', label='Stimulus Onset')

# Formatting
plt.title('Raster Plot of Neuronal Responses')
plt.xlabel('Time (s)')
plt.ylabel('Neuron')
plt.yticks(range(1, num_neurons + 1))
plt.legend()
plt.show()
```
